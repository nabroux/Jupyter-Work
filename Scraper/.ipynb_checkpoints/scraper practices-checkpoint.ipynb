# 標準 requests GET 請求

* 狀態碼 status_code 值為 200 就OK

```python
import requests

url = 'http://www.e-happy.com.tw'
html = requests.get(url)
html.encoding = 'utf-8'

if html.status_code == requests.codes.ok:
    print(html.text)
```

# 加上 URL 查詢參數

* 查詢參數定義為字典，加入GET請求中
* 用的是 params 參數

```python
import requests

payload = {'key1': 'value1', 'key2': 'value2'}
html = requests.get('http://httpbin.org/get', params=payload)
print(html.url)
```

# 自訂 HTTP Headers

* 自訂 user-agent

[各種瀏覽器字串連結](http://www.useragentstring.com/pages/useragentstring.php)

```python
url = 'http://www.e-happy.com.tw'

headers = {
    'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}

r = requests.get(url, headers=headers)
```

# POST 請求配參數

用的是 data 參數

```python
import requests

payload = {'key1': 'value1', 'key2': 'value2'}
html = requests.post('http://httpbin.org/post', data=payload)
print(html.text)
```

### 像 PTT 表特或八卦板都需要同意滿18，需要用cookie解決

```python
import requests
from bs4 import BeautifulSoup

# from 指從哪個網頁來， yes 是按鈕的預設值
payload = {
    'from': 'https://www.ptt.cc/bbs/Gossiping/index.html',
    'yes': 'yes'
}

headers = {
    'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}

rs = requests.Session()
rs.post('https://www.ptt.cc/ask/over18', data=payload, headers=headers)
res = rs.get('https://www.ptt.cc/bbs/Gossiping/index.html',
                  headers=headers)

soup = BeautifulSoup(res.text, 'html.parser')
items = soup.select('.r-ent')

for item in items:
    print(item.select('.date')[0].text,
         item.select('.author')[0].text,
         item.select('.title')[0].text)
```
